```python
column_names = [
    "x",
    "z",
    "r",
    "theta",
    "alpha",
    "psi",
    "q",
    "betak",
    "betat",
    "dbetatdr",
    "dbetatdt",
    "He",
    "Hf",
    "Omega",
    "b2",
    "rho",
]
df_2d = pd.read_csv(
    "/data/sjammi6/thesisproject/BlackHoleDiskProject/InitialData/routine3/data/2D_data/all_vars.dat", delimiter=r"\s+", header=None, names=column_names
)
df_2d
```

```python

import os
import pathlib 
dataframes = []
for f in pathlib.Path("/data/sjammi6/thesisproject/BlackHoleDiskProject/InitialData/routine3/data/grid_data").glob("*.d"):
    dataframes.append(str(f))
dataframes = [np.fromfile(f, dtype=np.float64) for f in dataframes]
dfs = []
failed = []
for d in dataframes:
    try:
        d = d[4:]
        d = d.reshape((-1, len(df_3d.columns)))
        d = pd.DataFrame(d, columns=list(df_3d.columns))
        dfs.append(d)
    except:
        failed.append(d)
df_interpolated = df_interpolated = pd.concat(dfs, ignore_index=True)
```

```python
print(len(failed), len(dfs))
df_interpolated = df_interpolated.map(lambda x: np.float64(0) if abs(x) <= 1e-10 else x)
df_interpolated
```

```python
df_interpolated_x = df_interpolated[(df_interpolated.z == np.float64(0)) & (abs(df_interpolated.y) <= np.float64(1e-10)) ]
```

```python
#    r_0 = np.sqrt(x**2 + y**2 + z**2)
#    theta_0 = np.arccos(z / r_0) if r_0 != 0 else np.float64(0)
#    phi_0 = np.arctan2(y, x) % (2 * np.pi)

# Calculate r
df_interpolated["r"] = np.sqrt(df_interpolated.x**2 + df_interpolated.y**2 + df_interpolated.z**2)

# Calculate theta
df_interpolated["theta"] = np.where(df_interpolated["r"] != 0, 
                                    np.arccos(df_interpolated.z / df_interpolated.r), 
                                    0)

# Calculate phi
df_interpolated["phi"] = np.arctan2(df_interpolated.y, df_interpolated.x) % (2 * np.pi)
```

```python
df_3d["r"] = np.sqrt(df_3d.x**2 + df_3d.y**2 + df_3d.z**2)

# Calculate theta
df_3d["theta"] = np.where(df_3d["r"] != 0, 
                                    np.arccos(df_3d.z / df_3d.r), 
                                    0)

# Calculate phi
df_3d["phi"] = np.arctan2(df_3d.y, df_3d.x) % (2 * np.pi)
```

```python
for col in list(df_3d.columns)[3:-3]:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(40, 10))
    ax1.set_xlim(1e-6, 100)
    ax1.set_ylim(1e-6, 100)
    ax2.set_xlim(1e-6, 100)
    ax2.set_ylim(1e-6, 100)
    ax1.set_yscale('log')
    ax1.set_xscale('log')
    tmpt1 = df_interpolated_xyplane.pivot(index='x', columns='y', values=col)
    m1 = ax1.pcolormesh(tmpt1, shading='auto', label="interpolated")
    plt.colorbar(m1, ax=ax1)
    ax1.set_title("{} interpolated - xy plane".format(col))
    #ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    tmpt2 = df_3d_xyplane.pivot(index='x', columns='y', values=col)
    m2 = ax2.pcolormesh(tmpt2, shading='auto', label="3d")
    plt.colorbar(m2, ax=ax2)
    ax2.set_title("{} 3d - xy plane".format(col))
    #ax2.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.savefig("{}_xyplane_3d_interpolated.png".format(col))
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

for col in list(df_3d.columns)[3:-3]:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

    # Interpolated data
    tmpt1 = df_interpolated_xyplane.pivot(index='x', columns='y', values=col)
    X1, Y1 = np.meshgrid(tmpt1.columns[:-1], tmpt1.index[:-1])
    m1 = ax1.pcolormesh(X1, Y1, tmpt1.values[:-1, :-1], shading='flat')
    plt.colorbar(m1, ax=ax1)
    ax1.set_title(f"{col} interpolated - xy plane")
    ax1.set_xlabel('Y')
    ax1.set_ylabel('X')

    # 3D data
    tmpt2 = df_3d_xyplane.pivot(index='x', columns='y', values=col)
    X2, Y2 = np.meshgrid(tmpt2.columns[:-1], tmpt2.index[:-1])
    m2 = ax2.pcolormesh(X2, Y2, tmpt2.values[:-1, :-1], shading='flat')
    plt.colorbar(m2, ax=ax2)
    ax2.set_title(f"{col} 3d - xy plane")
    ax2.set_xlabel('Y')
    ax2.set_ylabel('X')

    plt.tight_layout()
    plt.savefig(f"{col}_xyplane_3d_interpolated.png")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

for col in list(df_3d.columns)[3:-3]:
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

    # Interpolated data
    tmpt1 = df_interpolated_xyplane.pivot(index='x', columns='y', values=col)
    X1, Y1 = np.meshgrid(tmpt1.columns, tmpt1.index)
    m1 = ax1.pcolormesh(X1, Y1, tmpt1.values.T, shading='auto', label="interpolated")
    plt.colorbar(m1, ax=ax1)
    ax1.set_title(f"{col} interpolated - xy plane")
    ax1.set_xlabel('Y')
    ax1.set_ylabel('X')

    # 3D data
    tmpt2 = df_3d_xyplane.pivot(index='x', columns='y', values=col)
    X2, Y2 = np.meshgrid(tmpt2.columns, tmpt2.index)
    m2 = ax2.pcolormesh(X2, Y2, tmpt2.values.T, shading='auto', label="3d")
    plt.colorbar(m2, ax=ax2)
    ax2.set_title(f"{col} 3d - xy plane")
    ax2.set_xlabel('Y')
    ax2.set_ylabel('X')

    plt.tight_layout()
    plt.savefig(f"{col}_xyplane_3d_interpolated.png")
    plt.show()
```
